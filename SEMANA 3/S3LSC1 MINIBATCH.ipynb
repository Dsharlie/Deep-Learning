## 1. Comparacin de mtodos de lotes en el caso de regresin lineal simple

En este caso se trabajar un conjunto de datos simulado, usando como modelo generador de datos una regresin lineal con una sola variable predictora _input_. En particular se define el modelo:

$$ Y_i = \beta_0 + \beta_1X_i + \varepsilon_i $$

para $i=1,2,...,n=100$ y con valor de los parmetros $\beta_o=5$ y $\beta_1=1$. El error aleatorio se simula como  $\varepsilon_i\sim Normal(0,1)$, independiente para cada observacin.

### Generacin de Datos Simulados

A continuacin, se simulan los datos que se trabajarn en este notebook:

```python
import time
import numpy as np
from matplotlib import pyplot as plt

n=100
X = np.c_[[1]*n, 13.5 * np.random.rand(n, 1)]
y = 5 + X[:, 1].reshape((n, 1)) + np.random.randn(n, 1)
plt.scatter(X[:,1],y)
```

### Base comparativa: Estimacin por mnimos cuadrados

Se compararn las estimaciones realizadas por diferentes mtodos con la solucin analtica dada por mnimos cuadrados:

$$ \hat\beta= (X^TX)^{-1}X^TY $$

```python
XT=np.transpose(X)
beta_ols= np.dot(np.dot(np.linalg.inv(np.dot(XT,X)),XT),y)
print(beta_ols)
```

### Implementacion _Batch Gradient Descent_
Se usa toda la muestra para cada iteracin de descenso. Recordemos que, en este caso, el gradiente de la funcin de prdida (suma de cuadrados de errores) se calcula como:

$$ L(\beta)={2\over n} \left[X^TX\beta-X^TY\right] $$

En este caso, usaremos un _learning rate_ de $\alpha=0.001$.

```python
n_epochs = 50000   #Nmero mximo de pocas (ephocs) - en batch corresponden al nmero de iteraciones
beta = np.random.randn(2, 1)*0.05  #Inicializacin de parmetros
learning_rate=0.001
stop = False
epoch=0
loss_batch=list()
start = time.time()

while stop == False:
    gradients = (2/n) * (np.dot(np.dot(XT,X),beta) - np.dot(XT,y) )
    beta = beta - (gradients * learning_rate)
    if (np.linalg.norm(gradients) < 0.00001):
            stop = True
    if (epoch > n_epochs):
            stop = True
    epoch += 1
    loss_batch.append(1/n*np.dot((y-np.dot(X,beta)).T,(y-np.dot(X,beta))))
end = time.time()

print("Beta estimado: " +str(beta))
print("Tiempo de ejecucin: " +str(end - start))
print("Norma gradiente: " +str(np.linalg.norm(gradients)))
print("Nmero de iteraciones: " +str(epoch))
```

### Implementacion _Mini-Batch Gradient Descent_
En este caso se actualiza el gradiente usando un lote de tamao _mini_batch_size_ en cada iteracin.

Note que, por motivos de convergencia, se aumentan el nmero de _epochs_ y la tasa de aprendizaje; y se disminuye la tolerancia.


```python
n_epochs = 5000   #Nmero mximo de pocas (ephocs)
iterations = 1
learning_rate=0.005
mini_batch_size = 5
beta = np.random.randn(2, 1)*0.05  #Inicializacin de parmetros
stop = False
loss_mini=list()
start = time.time()

while stop == False:
    ini=int((iterations-1)%(n/mini_batch_size))*mini_batch_size
    Xb=X[ini:(ini+mini_batch_size):]
    yb=y[ini:(ini+mini_batch_size)]
    gradients = (2/n) * (np.dot(np.dot(Xb.T,Xb),beta) - np.dot(Xb.T,yb) )
    beta = beta - (gradients * learning_rate)
    if (np.linalg.norm(gradients) < 0.0005):  #Tenga en uenta que el gradiente cambia
            stop = True
    if (int(iterations/((n/mini_batch_size))) > n_epochs):
            stop = True
    iterations += 1
    loss_mini.append(1/n*np.dot((y-np.dot(X,beta)).T,(y-np.dot(X,beta))))
end = time.time()

print("Beta estimado: " +str(beta))
print("Tiempo de ejecucin: " +str(end - start))
print("Norma gradiente: " +str(np.linalg.norm(gradients)))
print("Nmero de epochs: " +str(int(iterations/((n/mini_batch_size)))))
```

El caso de la versin compltamente estocstica _Stochastic Gradient Descent_ (SGD), equivale a hacer el tamao de cada lote igual a 1.

## 2. Implementacin de mini-_batch_ en _Keras_ y comparacin de tamaos de lote

En _Keras_ el parmetro usado para controlar el tamao de lote es: "batch_size" Dependiendo de las configuraciones de mquina particulares, _Keras_ usa por _default_ la estrategia de mini-_batch_, con un tamao de lote de 32. 

En este caso, uasaremos la base de datos del _National Institute of Diabetes and Digestive and Kidney Diseases_. La variable de inters 'Clase' es una variable binaria que toma el valor de 1 en caso de que el paciente tenga diabetes o de 0 en caso contrario. Estos datos ya haban sido usados en un ejercicio anterior y en este caso, exploraremos la sensibilidad de la estimacin de la red en trminos del _batch_.

```python
import pandas as pd
from numpy import loadtxt
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn import metrics

dataset = pd.read_csv('test2.csv', sep=",")
dataset=dataset.to_numpy()
X_train, X_test, y_train, y_test = train_test_split(dataset[:,0:8], dataset[:,8], test_size=0.20)
```

Ahora estimamos el modelo. Primero usaremos la estrategia _batch_ , es decir, se usan todos los datos en cada iteracin del algoritmo.

```python
n=X_train.shape[0]

model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
start = time.time()

history=model.fit(X_train, y_train, epochs=250, batch_size=n)
end = time.time()
```

```python
plt.plot(history.history['loss'])
plt.title('Funcin de prdida del modelo')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()

print("Tiempo de ejecucin: " +str(end - start))
```

### Tamao de lote intermedio:
